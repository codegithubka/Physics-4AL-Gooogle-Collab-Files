```python
# Run this cell to mount your Google Drive.
from google.colab import drive
drive.mount('drive')

time = []
displacement = []
```

```python
import numpy as np
import csv
import matplotlib.pyplot as plt
with open('/content/drive/MyDrive/PHYSICS4AL/Unit 3 - Oscillations/3D/Copy of 43 - Sheet1.csv', newline="") as f: 
  csvreader = csv.reader(f) 
  for row in csvreader: 
    time.append(float(row[0]))
    displacement.append(float(row[1]))
```

```python
newList = []
for x in time:
    newList.append(x/1000)
newL = []
for y in displacement:
    newL.append(y/100)

plt.plot(newList,newL)
plt.title('Position Data for Oscillator at Resonance')
plt.ylabel('Displacement (m)')
plt.xlabel('Time (s)')
```
