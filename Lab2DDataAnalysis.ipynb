```python
import csv
import array
from statistics import stdev

array1 = []
array2 = []
array3 = []

with open("/content/drive/MyDrive/z_axis_labdata.csv", newline="") as f: 
  csvreader = csv.reader(f) 
  for row in csvreader: 
    i = float(row[2])
    j = float(row[3])
    k = float(row[4])
    print(i, j, k)

    array1.append(i)
    array2.append(j)
    array3.append(k)

print(array1)
print(array2)
print(array3)
    

 
```

```python
import statistics
third_row_stdev = statistics.stdev(array1)
third_row_mean = statistics.mean(array1)

fourth_row_stdev = statistics.stdev(array2)
fourth_row_mean = statistics.mean(array2)

fifth_row_stdev = statistics.stdev(array3)
fifth_row_mean = statistics.stdev(array3)

print(third_row_mean, fourth_row_mean, fifth_row_mean, '\n')

print(third_row_stdev, fourth_row_stdev, fifth_row_stdev)



```
